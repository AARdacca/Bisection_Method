```python
import numpy as np
import matplotlib.pyplot as plt

# The function returns when a value xm is found for which the absolute relative approximate error goes under tolerance.
def arae(x_prev, x_pres):
	return abs((x_pres - x_prev)/x_pres) * 100

def plot(f, solution, init_guess_0, init_guess_1, find="Nothing", newfirst=0.0, newsecond=0.0):
	x = np.linspace(init_guess_0-1, init_guess_1+1, 10000)
	y = f(x)
	
	plt.plot(x, y, label = 'f(x)', linewidth=3)
	plt.axhline(y=0, color='k')
	if init_guess_0 > 0:
		pass
	else:
		plt.axvline(x=0, color='k')
	plt.axvline(x=solution, color='g', linestyle='-', label='$solution=((first guess)+(second guess))/2$')
	plt.axvline(x=init_guess_0, color='lightcoral', linestyle='--', label='$first guess$')
	plt.axvline(x=init_guess_1, color='firebrick', linestyle='--', label='$second guess$')
	if find == "Nothing":
		pass
	else:
		plt.axvline(x=float(find), color='forestgreen', linestyle='-.', label='$mid$')
		plt.axvline(x=newfirst, color='purple', linestyle='--', label='$new first guess$')
		plt.axvline(x=newsecond, color='violet', linestyle='--', label='$new second guess$')
	plt.grid()
	plt.legend()
	print(f"This is for {f.__name__}.")
	plt.show()
	
```

**Task A**
Implement the Bisection method algorithm to solve a non-linear polynomial equation. Your
function must take the following as parameters:
~~----------------------------------------------------------------------------------------------------~~
**f : function**
The function for which we are trying to find a root.
**tolerance : number**
The function returns when a value xm is found for which the absoluterelative approximate error goes under tolerance.
**maxiter : integer**
If the function exceeds maxiter iterations, then it will return None.

```python
# Python program for implementation
# of Bisection Method for
# solving equations
# ------------------------------------------------------------------------
# Prints root of func(x)
def bisection(f, first_guess, second_guess, tolerance, maxiter):
	arr1=[]; arr2=[]; arr3=[]; arr0=[];
	if f(first_guess) * f(second_guess) >= 0:
		print("You have not assumed right first_guess and second_guess\n")
		return
	iteration = 0
	mid_value = first_guess
	# while (second_guess - first_guess) >= 0.000000005:
	while True:
		# Previous mid value
		prev_mid_value = mid_value
		# Here runs iteration breaker
		if iteration == maxiter:
			arr0.append(arr1); arr0.append(arr2); arr0.append(arr3);
			break
		# Find middle point
		mid_value = (first_guess + second_guess) / 2
		arr1.append(mid_value); arr2.append(first_guess); arr3.append(second_guess);
		# Check tolerance
		# check = arae(prev_mid_value, mid_value)
		# print(check)
		if (arae(prev_mid_value, mid_value))<=tolerance:
			arr0.append(arr1); arr0.append(arr2); arr0.append(arr3);
			break
		# Check if middle point is root
		if f(mid_value) == 0.0:
			arr0.append(arr1); arr0.append(arr2); arr0.append(arr3);
			break
		# Decide the side to repeat the steps
		if f(mid_value)*f(first_guess) < 0:
			second_guess = mid_value
		else:
			first_guess = mid_value
		# iteration increase
		iteration = iteration+1
	print(f"The value of {f.__name__} root is : {mid_value:.7f}")
	arr0.append(mid_value)
	return arr0
```

**Task B**
Use this algorithm to find the solution for the following equations:
● x^2^ − 5x + 6 = 0
● x^3^ − 3x^2^ + 3 = 0
● x^4^ − 2x^3^ + x = 0

```python
# The function for which we are trying to find a root.
def function1(x):
    return x**2 - 5*x + 6
def function2(x):
	return x**3 - 3*x**2 + 3
def function3(x):
	return x**4 - 2*x**3 + x

# Driver code
# Initial values assumed
a = 2.5; b = 5
# Above variables using x = -b/2a where x = a for this and an is the coefficient of the x^2 term (1 in this case). b is the coefficient of the x term (-5 in this case).
# Or a>2 and a<3 
c = -20; d = 30
# Above variables are assumed using wild guess
e = 1.5;  f = 2
# Above 'e' variable has the expression x^4−2x^3+x which is negative for x in the intervals −∞,−1 and (−1,1).

p = bisection(function1, a, b, 0.001, 100)
q = bisection(function2, c, d, 0.001, 20)
r = bisection(function3, e, f, 0.001, 100)
```

## Plotting

```python
# Plotting the solution.
plot(function1, p[3], a, b)
plot(function2, q[3], c, d)
plot(function3, r[3], e, f)
# print(p, q, r, sep="\n\n")
```

## Iterations

```python
# This is iterations
def iteration_plot(func, solution, init_guess_0, init_guess_1, arr1, arr2, arr3):
	n=len(arr1); print(f"The Number of steps is {n}")
	# print(arr1, arr2, arr3)
	for i in range(0, n):
		plot(func, solution, init_guess_0, init_guess_1, str(arr1[i]), arr2[i], arr3[i])
```
